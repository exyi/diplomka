```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import polars as pl
import seaborn as sns
```

```python
df_fr3d = pl.read_parquet("/home/exyi/Downloads/basepairs/cWH-G-G-filtered.parquet")
df_all = pl.read_parquet("/home/exyi/Downloads/basepairs/cWH-G-G-filtered-allcontacts.parquet")
```

```python
fig, [ax, ax2] = plt.subplots(1, 2, figsize=(10, 5))

ax.set_xlabel('Yaw angle (left to right, degrees)')
ax.set_ylabel('Count')
ax.set_title('Yaw ')

sns.histplot(df_all["C1_C1_yaw1"], bins=np.linspace(-180, 180, 60), ax=ax, color='lightblue', label='All')
sns.histplot(df_fr3d["C1_C1_yaw1"], bins=np.linspace(-180, 180, 60),  ax=ax, color='red', label='FR3D')

ax.legend([
    f"All G-G contacts ({len(df_all)})",
    f"FR3D cWH G-G ({len(df_fr3d)})"
])

ax=ax2

ax.set_xlabel('N1 · · · O6 H-bond length (Å)')
ax.set_ylabel('Count')
ax.set_title('Hydrogen bond length')

sns.histplot(df_all["hb_0_length"], bins=np.linspace(2, 6, 60), ax=ax, color='lightblue', label='All')
sns.histplot(df_fr3d["hb_0_length"], bins=np.linspace(2, 6, 60),  ax=ax, color='red', label='FR3D')

ax.legend([
    f"All G-G contacts ({len(df_all)})",
    f"FR3D cWH G-G ({len(df_fr3d)})"
])

fig.savefig("cWH-G-G-yaw-hbond.svg")
```
