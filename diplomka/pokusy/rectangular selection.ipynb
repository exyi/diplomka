```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as patches
```

```python
x_ = np.random.sample(200)
y_ = np.random.sample(x_.size)
x = x_ + 1/2 * (y_*2) ** 2 - 0.5
y = y_ - 0.5

plt.scatter(x, y)

rect = [
    np.min(x),
    np.min(y),
    np.max(x),
    np.max(y)
]

nx, ny = np.random.normal(np.mean(x), np.std(x), size=50), np.random.normal(np.mean(y), np.std(x), size=50)
plt.scatter(nx, ny)

plt.gca().add_patch(patches.Rectangle((rect[0]-0.05, rect[1]-0.05), rect[2] - rect[0] + 0.1, rect[3] - rect[1] + 0.1, linewidth=1, edgecolor='r', facecolor='none'))
```
